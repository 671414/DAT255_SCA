This notebook is used to visualize traces, and can be used to troubleshoot datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
import src.DAT255_SCA.utils.dataset_utils as utils
import os


ATTACK_BYTE = 0
ATTACK_POINT = ("sub_bytes_in")
NUM_TRACES = 10
val_traces = []
train_traces = []
test_traces = []
```

```python
#looping through the datasets and fetching first trace
dir = os.fsencode("../datasets/attack/")
for file in os.scandir(dir):
    filename = os.fsdecode(file)
    print(filename)
    if filename.endswith(".hdf5"):

        x,y,keys,pts = utils.load_and_prepare_dataset_for_evaluation(filename, ATTACK_BYTE, ATTACK_POINT, num_traces=NUM_TRACES)
        val_traces.append(x[0][0])
```

```python
file_path = "../datasets/train/training_0.hdf5"
x, y, keys, pts = utils.load_and_prepare_dataset_for_evaluation(file_path, ATTACK_BYTE, ATTACK_POINT, num_traces=NUM_TRACES)
train_traces.append(x[0][0])
```

```python
file_path = "../datasets/test/testing_0.hdf5"
x, y, keys, pts = utils.load_and_prepare_dataset_for_evaluation(file_path, ATTACK_BYTE, ATTACK_BYTE, num_traces=NUM_TRACES)
test_traces.append(x[0][0])
```

```python
plt.figure(figsize=(25, 8))
#plt.plot(val_traces[5][0:150])
#plt.plot(traces[1][0:150])
#plt.plot(traces[2][0:150])
#plt.plot(traces[3][0:150])
plt.plot(val_traces[6][0:150], label = "validation")
plt.plot(train_traces[0][0:150], label = "training")
plt.title("traces", fontsize=20)
plt.legend(fontsize=30, loc='upper left', bbox_to_anchor=(1, 1))
plt.tick_params(axis='both', which='major', labelsize=14, width=2, length=6)  # Tykkere ticks
plt.tick_params(axis='both', which='minor', length=4, width=1.5)
```
